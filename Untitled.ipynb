```python
import numpy as np
import matplotlib.pyplot as plt
import pywt # 웨이블릿 변환 라이브러리 

plt.rcParams.update({'figure.max_open_warning': 0})
# 이미지 크기 설정
plt.rcParams['figure.figsize'] = 12, 6

ecgdata = np.loadtxt('./eeg/S001.csv', delimiter=',', dtype=np.float32)
for i in range(int(len(ecgdata) / 512)):
    coeffs = pywt.wavedec(ecgdata[i * 512: (i + 1) * 512], 'db4', level=4)
    cA4, cD4, cD3, cD2, cD1 = coeffs # 웨이블릿 변환한 특징 값들이  저장됨 
#   16   32   64   128 256   
# 웨이블릿 변환한 값이 cA4.. 인데 각각 평균을 뽑아 

# cA4, cD4 ... 의 값들에 특징을 각각 n개씩 뽑음.   

    cD3x = []
    s = 1
    for j in range(int(len(cD3))):
        cD3x.append(s)
        s = s + 1

    cD4x = []
    s = 1
    for j in range(int(len(cD4))):
        cD4x.append(s)
        s = s + 1

    #cD = []

    fig = plt.figure()

    # 1 X 2 형태로 2개의 이미지를 표시 fig.add_subplot(1, 2, 1), fig.add_subplot(1, 2, 2)
    # 2 X 2 형태로 4개의 이미지를 표시 fig.add_subplot(2, 2, 1), fig.add_subplot(2, 2, 2), fig.add_subplot(2, 2, 3), fig.add_subplot(2, 2, 4)
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    #ax3 = fig.add_subplot(2, 2, 3)
    #ax4 = fig.add_subplot(2, 2, 4)

    ax1.plot(cD3x, cD3)
    ax2.plot(cD4x, cD4)
    #ax3.plot(cD, cD)
    #ax4.plot(cD4x, cD4)

    plt.show()

    #plt.plot(x,cD2,'black')
    #plt.xlabel('time')
    #plt.ylabel('mV')
    #plt.show()
```
